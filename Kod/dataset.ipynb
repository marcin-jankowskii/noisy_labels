```python
import os
import cv2
import numpy as np

# Ścieżki do folderów z obrazami binarnymi
folder1_path = 'D:\\Programowanie\\Python\\Magisterka\Praca Dyplomowa\\noisy_labels\\Dataset\\SegSperm\\test\\GT1_full'
folder2_path = 'D:\\Programowanie\\Python\\Magisterka\Praca Dyplomowa\\noisy_labels\\Dataset\\SegSperm\\test\\GT1_full\\tail'
output_path = 'D:\\Programowanie\\Python\\Magisterka\Praca Dyplomowa\\noisy_labels\\Dataset\\SegSperm\\test\\GT1_full\\head'


# Utwórz folder z różnicą, jeśli nie istnieje
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Funkcja do wczytywania i zapisywania obrazów binarnych
def load_binary_image(file_path):
    image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
    return image

# Lista plików w pierwszym i drugim folderze
files1 = os.listdir(folder1_path)
files2 = os.listdir(folder2_path)

# Porównaj obrazy i zapisz różnice
for file in files1:
    if file in files2:
        image1 = load_binary_image(os.path.join(folder1_path, file))
        image2 = load_binary_image(os.path.join(folder2_path, file))

        # Oblicz różnicę między obrazami
        difference = cv2.absdiff(image1, image2)

        # Zapisz obraz różnicy w folderze wyjściowym z tą samą nazwą
        output_file_path = os.path.join(output_path, file)
        cv2.imwrite(output_file_path, difference)

print("Zakończono zapisywanie obrazów z różnicą.")
```

```python
import os

# Ścieżka do folderu, w którym znajdują się pliki PNG
folder_path = "D:/Programowanie/Python/Magisterka/Praca Dyplomowa/noisy_labels/Dataset/SegSperm/test/images/"

# Pobierz listę plików w folderze
png_files = [f for f in os.listdir(folder_path) if f.endswith(".png")]

# Sortuj pliki alfabetycznie, aby zachować pierwotny porządek
png_files.sort()

# Licznik numerów plików
file_counter = 0

# Iteruj przez pliki i zmieniaj ich nazwy
for filename in png_files:
    # Przygotuj nową nazwę
    new_filename = f"{file_counter}.png"

    # Pełna ścieżka do starego pliku
    old_filepath = os.path.join(folder_path, filename)

    # Pełna ścieżka do nowego pliku
    new_filepath = os.path.join(folder_path, new_filename)

    # Zmień nazwę pliku
    os.rename(old_filepath, new_filepath)

    print(f"Zmieniono nazwę {filename} na {new_filename}")

    # Zwiększ licznik
    file_counter += 1

print("Zmieniono nazwy wszystkich plików.")
```
